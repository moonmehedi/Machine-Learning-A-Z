```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:\ml\resorces\Machine Learning-A-Z-Codes-Datasets\Machine Learning A-Z (Codes and Datasets)\132 - Machine-Learning-A-Z-Model-Selection\Machine Learning A-Z (Model Selection)\Regression\Data.csv")
data
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
y = y.reshape(-1, 1)
x, y
```

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
sc_y = StandardScaler()

x_scaled = sc_x.fit_transform(x)
y_scaled = sc_y.fit_transform(y)

x_scaled, y_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x_scaled, y_scaled, test_size=.1, random_state=0)
x_train, x_test, y_train, y_test
```

```python
from sklearn.svm import SVR

regressor = SVR(kernel='rbf')

regressor.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(sc_y.inverse_transform(y_test), sc_y.inverse_transform(
    regressor.predict(x_test).reshape(-1, 1)))
```
