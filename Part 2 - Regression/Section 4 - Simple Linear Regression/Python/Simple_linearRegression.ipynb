IMPORTING THE LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IMPORT THE DATA SET

```python
dataSet=pd.read_csv(r"D:\ml\resorces\Machine Learning-A-Z-Codes-Datasets\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 4 - Simple Linear Regression\Python\Salary_Data.csv")
x=dataSet.iloc[:,:-1].values
y=dataSet.iloc[:,-1].values
y
```

SPLITTING THE DATASET INTO TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=(train_test_split(x,y,test_size=0.2,random_state=0))
y_test
```

TRAIN THE MODE

```python
from sklearn.linear_model import LinearRegression  #here sklearn.linear_mode is a library and LinearRegression is a class
regressor= LinearRegression()
regressor.fit(x_train,y_train)
```

PREDICT THE RESULT

```python
y_predict= regressor.predict(x_test,)
```

visualization

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,regressor.predict(x_train),color='red') 
plt.title('regression model to predict salary accoring to experiance')
plt.xlabel('years of expericane')
plt.ylabel('Predicted salary')

#First, we have trained our regression model now if we want to drow the regression line with our training data we give train data as test case and for training data it predicts the values
```

TEATING WITH TEST CASES

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,regressor.predict(x_test),color='red')
plt.title('regression model to predict salary accoring to experiance')
plt.xlabel('years of expericane')
plt.ylabel('Predicted salary')
```
