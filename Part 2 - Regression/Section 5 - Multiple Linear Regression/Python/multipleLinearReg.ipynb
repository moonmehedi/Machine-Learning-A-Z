```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import data

startUps=pd.read_csv(r"D:\ml\resorces\Machine Learning-A-Z-Codes-Datasets\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 5 - Multiple Linear Regression\Python\50_Startups.csv")
startUps['isNewYork']=(startUps['State']=='New York').astype(int)
startUps['isCalifornia']=(startUps['State']=='California').astype(int)
p=startUps['Profit']
startUps=startUps.drop(['State','Profit'],axis=1)
startUps["Profit"]=p


startUps
```

```python
#import data

x=startUps.iloc[:, :-1].values
y=startUps.iloc[:, -1].values
x

```

```python
#standardization of y

from sklearn.preprocessing import MinMaxScaler

scaller=MinMaxScaler()
x=scaller.fit_transform(x)
x
```

```python

#split data into train and test set

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)

x_train
```

```python
#import model

from sklearn .linear_model import LinearRegression

regressor=LinearRegression()
regressor.fit(x_train,y_train)


```

```python
#examine data


y_pred=regressor.predict(x_test)
np.set_printoptions(precision=2)
arr= np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
arr
```
