```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'Social_Network_Ads.csv')

x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

x=scaler.fit_transform(x)

x
```

```python
from sklearn.model_selection import train_test_split

xTrain,Xtest,Ytrain,Ytest=train_test_split(x,y,test_size=.2,random_state=0)
xTrain,Xtest,Ytrain,Ytest
```

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression()

regressor.fit(xTrain,Ytrain)
```

```python

np.concatenate((Ytest.reshape(-1,1),regressor.predict(Xtest).reshape(-1,1)),1)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(Ytest,regressor.predict(Xtest).reshape(-1,1))
accuracy_score(Ytest,regressor.predict(Xtest).reshape(-1,1))
```
