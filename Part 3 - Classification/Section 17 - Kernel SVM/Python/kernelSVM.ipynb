```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv(r"Social_Network_Ads.csv")

x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler

sc_x=StandardScaler()

x_scaled=sc_x.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=.25,random_state=0)

x_train, x_test,y_train,y_test
```

```python
from sklearn.svm import SVC

classifier=SVC()

classifier.fit(x_train,y_train)

y_pred=classifier.predict(x_test)
y_pred
```

```python
np.concatenate((y_test.reshape(-1,1),y_pred.reshape(-1,1)),axis=1)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

cm= confusion_matrix(y_test.reshape(-1,1),y_pred.reshape(-1,1))

print(cm)

accuracy_score(y_test,y_pred)
```
