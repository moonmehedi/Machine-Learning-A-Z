```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'Mall_Customers.csv')
data
```

```python
data.describe()
```

```python

ColumnOfFigure=data.columns[1:-1]


fig,axis=plt.subplots(nrows=len(ColumnOfFigure),ncols=1,figsize=(24,24))


for i,column in enumerate(ColumnOfFigure):
    axis[i].scatter(data.iloc[:,i+1],data['Spending Score (1-100)'])
    axis[i].set_xlabel(column)
    axis[i].set_ylabel('spending')
```
